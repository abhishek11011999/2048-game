```python
import random # to generate random number
#whenever start a game then 4*4 matrix will be created with all zero value
def start_game():
    mat=[]
    for i in range(4):
        mat.append([0]*4)
    return mat

#add 2 at random position at empty postion
#index of matrix go from [0 to 3]
def add_new_2(mat):
    r=random.randint(0,3) # to generate random poition for row 
    c=random.randint(0,3) # to generate random position for column
    while(mat[r][c]!=0): # we keep genearting random number until not find empty place
        r=random.randint(0,3)
        c=random.randint(0,3)
    mat[r][c]=2 #now add 2 to empty place

#best approach to combine two tiles is first compress the matrix 
# in compress all non-zero number are comes one side
# we make new matrix and fill each row only for non -zero value i.e increase i only see non-zero value on old matrix
#e.g initially matrix 
#[2 0 0 2]
#[2 0 4 4]
#[2 2 2 2]
#[4 4 0 0]

#after compress
#[2 2 0 0]
#[2 4 4 0]
#[2 2 2 2]
#[4 4 0 0]


'''
def compress(mat):
    new_mat=[]
    for i in range(4):
        new_mat.append([0]*4)
    for i in range(4):
        pos=0
        for j in range(4):
            if mat[i][j]!=0:
                new_mat[i][pos]=mat[i][j]
                pos+=1
    return new_mat
'''

# modified compress function by adding changed used in each move

def compress(mat):
    changed=False
    new_mat=[]
    for i in range(4):
        new_mat.append([0]*4)
        
    for i in range(4):
        pos=0
        for j in range(4):
            if mat[i][j]!=0:
                new_mat[i][pos]=mat[i][j] # while copy if changed happend then pos and j position will be diff and vice versa
                if j!=pos:
                    changed=True
                pos+=1
    return new_mat,changed



#now we have to merge the two tiles concesutive tiles 
#e.g initially matrix 
#[2 2 4 0]
#[4 4 4 4]
#[2 0 0 0]
#[4 2 2 0]

#after merge
#[4 0 4 0]
#[8 0 8 0]
#[2 0 0 0]
#[4 4 0 0]
'''
def merge(mat):
    for i in range(4):
        for j in range(3): #only go upto second last column
            if mat[i][j]==mat[i][j+1] and mat[i][j]!=0:
                mat[i][j]=mat[i][j]*2
                mat[i][j+1]=0
    return mat
'''

# modified merge function by adding changed used in each move
def merge(mat):
    changed=False
    for i in range(4):
        for j in range(3): #only go upto second last column
            if mat[i][j]==mat[i][j+1] and mat[i][j]!=0:
                mat[i][j]=mat[i][j]*2
                mat[i][j+1]=0
                changed=True  # as merge happens
    return mat,changed



# Now we have to reverse each row of matrix
#e.g initially matrix 
#[2 2 4 0]
#[0 4 4 2]
#[2 2 4 4]
#[4 2 4 2]

#after merge
#[0 2 4 2]
#[2 4 4 0]
#[4 4 2 2]
#[2 4 2 4]

def reverse(mat):
    new_mat=[]
    for i in range(4):
        new_mat.append([]) # for each row append a list in new matrix
        for j in range(4):
            new_mat[i].append(mat[i][4-j-1]) # at 0th index of new mat we append 3rd (4-0-1) ele of mat.
    
    return new_mat


def transpose(mat):
    new_mat=[]
    for i in range(4):
        new_mat.append([])
        for j in range(4):
            new_mat[i].append(mat[j][i])
    return new_mat


#All moves========================

# left moves: 1) deals with j and j+1
# step-1: Compress
# Step-2: Merge
# Step-3: Compress

# Right moves: 1) deals with j and j-1
# step-1: Compress
# Step-2: Merge
# Step-3: Compress
# OR (use same logic as of left move logic)
# Step-1: Reverse
# step-1: Compress (same as left logic)
# Step-2: Merge (same as left logic) -----now deals with j and j+1
# Step-3: Compress (same as left logic)
#step-4: Reverse

# UP moves: 1) deals with i and i+1
# step-1: Compress
# Step-2: Merge
# Step-3: Compress
# OR (use same logic as of left move logic)
# Step-1: Transpose
# step-1: Compress (same as left logic)
# Step-2: Merge (same as left logic) ----now deals with j and j+1
# Step-3: Compress (same as left logic)
#step-4: Transpose


# DOWN moves: 1) deals with i and i+1
# Step-1: Transpose (after this apply same logic as right logic)
# Step-2: Reverse
# step-3: Compress (same as left logic)
# Step-4: Merge (same as left logic) -----now deals with j and j+1
# Step-5: Compress (same as left logic)
# Step-6: Reverse
# step-4: Transpose

#==================================================
# we call add_new_2 only when any change is happens from prev state 
#change means 1) atleast one block tile movement( any compress) or 2) merging happens for atleast one time

# now we have to chek change in every moves.
# so ( 1st compress and merge ) change anythings then only 2nd compress work make change anythings so we have to chack only first two fun


def move_left(grid):
    new_grid,changed1=compress(grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2 # changed=[True or False]
    new_grid=compress(new_grid)
    return new_grid,changed

def move_right(grid):
    reversed_grid=reverse(grid)
    new_grid,changed1=compress(reversed_grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid=compress(new_grid)
    final_grid=reverse(new_grid)
    return final_grid,changed

def move_up(grid):
    transposed_grid= transpose(grid)
    new_grid,changed1= compress(transposed_grid)
    new_grid,changed2= merge(new_grid)
    changed=changed1 or changed2
    new_grid= compress(new_grid)
    final_grid= transpose(new_grid)
    return final_grid,changed

def move_down(grid):
    transposed_grid= transpose(grid)
    reversed_grid=reverse(transposed_grid)
    new_grid,changed1= compress(reversed_grid)
    new_grid,changed2= merge(new_grid)
    changed=changed1 or changed2
    new_grid= compress(new_grid)
    final_reversed_grid=reverse(new_grid)
    final_grid= transpose(final_reversed_grid)
    return final_grid,changed


#three game state: 1) WON 2) LOST 3) GAME NOT OVER ,so my function return these three state
def get_game_state(mat):
    # Antwhere 2048 is present
    for i in range(4):
        for j in range(4):
            if(mat[i][j]==2048):
                return 'WON'
    
    
    #Anywhere 0 is present then we are still in game
    for i in range(4):
        for j in range(4):
            if(mat[i][j]==0):
                return 'GAME NOT OVER'
    
    # if whole mat is filled ,but still we are in game as if any consecutive pos are equal then we can move in that direction and combine the tiles
    # for each tiles at (i,j) we just check its right (i,j+1) and its down (i+1,j) as by doing so we covers all consecutive
    # now by doing so we checked in range [0 to 2] index only as at index 3-column (i,j+1) and at index 3-row (i+1,j), is not valid
    
    #Every row and column except last row and last column
    for i in range(3):
        for j in range(3):
            if(mat[i][j]==mat[i+1][j] or mat[i][j]==mat[i][j+1]):
                return 'GAME NOT OVER'
            
    # now we have to check espically for last row and col like last row [2,4,4,2] and last col [2,2,2,2]
    
    # for last row 
    for j in range(3):
        if(mat[3][j]==mat[3][j+1]):
            return 'GAME NOT OVER'
    
    #last column
    for i in range(3):
        if(mat[i][3]==mat[i+1][3]):
            return 'GAME NOT OVER'
    
    return 'LOST' # if neither of them is satisfied







    
```

```python
from tkinter import *

import LogicsFinal
import constants as c

class Game2048(Frame): #our Game2048 inheritate thr super class Frame
    def __init__(self):
        Frame.__init__(self) #constructor
        self.grid() #now frame look like grid now
        self.master.title('2048') # everything is title as 2048, 
        self.master.bind('<Key>',self.key_down) # if any key presses then call kry_down function
        self.commands={c.KEY_UP:LogicsFinal.move_up,c.KEY_DOWN:LogicsFinal.move_down,c.KEY_LEFT:LogicsFinal.move_left,c.KEY_RIGHT:LogicsFinal.move_right}
        
        self.grid_cells=[] # to add cell in grid
        self.init_grid() # Add the grid cells
        self.init_matrix() #fun= init_matrix
        self.update_grid_cells() # this fun changed when anythings happens, basicaaly it changes UI
        self.mainloop()  # actually run the program

    def init_grid(self): #it make grid
        background=Frame(self,bg=c.BACKGROUND_COLOR_GAME,width=c.SIZE,height=c.SIZE) # aframe was created of size 400*400 inside outer frame
        background.grid() # make frame as grid
        for i in range(c.GRID_LEN):#GRID_LEN=4
            grid_row=[]
            for j in range(c.GRID_LEN):# for each row i here we add cell
                cell=Frame(background,bg=c.BACKGROUND_COLOR_CELL_EMPTY,width=c.SIZE/c.GRID_LEN,height=c.SIZE/c.GRID_LEN) # cell is another frame inside the inner background
                cell.grid(row=i,column=j,padx=c.GRID_PADDING,pady=c.GRID_PADDING) # make cell frame attached to grid of i, j grid with added padding
                t=Label(master=cell,text="",bg=c.BACKGROUND_COLOR_CELL_EMPTY,justify=CENTER,font=c.FONT,width=5,height=2) # label created inside cell with text at centre act as number input
                t.grid() # so inside label color is change we number change so we make label as grid
                grid_row.append(t) #and attached to grid
            self.grid_cells.append(grid_row) #now cell is attached grid
        
    def init_matrix(self): #it make matrix
        self.matrix=LogicsFinal.start_game() #the matrix withh all zero will reflected back into game
        LogicsFinal.add_new_2(self.matrix) # 
        LogicsFinal.add_new_2(self.matrix)
    
    def update_grid_cells(self): # it update matrix into grid
        for i in range(c.GRID_LEN):
            for j in range(c.GRID_LEN):
                new_number=self.matrix[i][j]
                if new_number==0:
                    self.grid_cell[i][j].configure(text="", bg=c.BACKGROUND_COLOR_CELL_EMPTY)
                else:
                    self.grid_cell[i][j].configure(text=str(new_number),bg=c.BACKGROUND_COLOR_DICT[new_number],fg=c.CELL_COLOR_DICT[new_number]) #fg=foreground
        self.update_idletasks() #it wait until all color changed
    
    def key_down(self,event):
        key=repr(event.char) # which event is happens like w is pressed or a is pressed etc
        if key in self.commands:# if that key exist in command or not
            self.matrix,changed=self.commands[repr(event.char)](self.matrix)
            if changed:
                LogicsFinal.add_new_2(self.matrix)
                self.update_grid_cells()
                changed=False
                if LogicsFinal.get_game_state(self.matrix)=='WON':
                    self.grid_cell[1][1].configure(text="YOU",bg=c.BACKGROUND_COLOR_CELL_EMPTY)
                    self.grid_cell[1][2].configure(text="WIN",bg=c.BACKGROUND_COLOR_CELL_EMPTY)
                if LogicsFinal.get_game_state(self.matrix)=='LOST':
                    self.grid_cell[1][1].configure(text="YOU",bg=c.BACKGROUND_COLOR_CELL_EMPTY)
                    self.grid_cell[1][2].configure(text="LOSE!",bg=c.BACKGROUND_COLOR_CELL_EMPTY)   

gamegrid=Game2048
```
